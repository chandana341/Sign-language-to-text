```python
# Importing the required libraries

import cv2
import numpy as np
import os
import string
```

```python
# Creating the directory Structure "dataset"

if not os.path.exists("dataSet"):
    os.makedirs("dataSet")

if not os.path.exists("dataSet/testingData"):
    os.makedirs("dataSet/testingData")


# Making folder  0-9 (blank) in the testing data folder
for i in range(10):
    if not os.path.exists("dataSet/testingData/" + str(i)):
        os.makedirs("dataSet/testingData/" + str(i))

# Making Folders from A to Z in the testing data

for i in string.ascii_uppercase:
    if not os.path.exists("dataSet/testingData/" + i):
        os.makedirs("dataSet/testingData/" + i)
```
